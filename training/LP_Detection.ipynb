```python
results = model.train(
    data=yaml_file,
    imgsz=320,
    batch=16,
    epochs=30,
    workers=4,
    project="runs/train",
    name="LP_detection",
    exist_ok=True
)
```

```python
from ultralytics import YOLO

# Load your trained YOLOv8 model
model = YOLO("runs/train/LP_detection/weights/best.pt")

# Path to your test image
test_image = "data/lp/images/val/xemay1209.jpg"

# Run inference (returns a list of Results)
results = model.predict(source=test_image, imgsz=320, conf=0.25)

# Access the first result
result = results[0]

# Show the image with bounding boxes
result.show()

# Save the annotated image to a folder
result.save("runs/test")
```
